## Logistic Regression을 사용해서 와인 데이터 분류

- 예측값이 0.5보다 작을 경우 0, 0.5보다 클 경우 1로 예측(이진 분류)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
wine=pd.read_csv("wine.csv")
wine
```

```python
#del wine['Unnamed: 0']
x=wine.iloc[:,1:]
y=wine['type']
```

```python
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LogisticRegression  # 이 모듈이 로지스틱 회귀모형과 관련된 모듈
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
lr=LogisticRegression(C=1,max_iter=1000)
lr.fit(x_train,y_train)
train_score=lr.score(x_train,y_train)
test_score=lr.score(x_test,y_test)
print("train score : ",train_score)
print("test score : ",test_score)
print("train score - test score : ",train_score-test_score)
```
