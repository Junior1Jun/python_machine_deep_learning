```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
```

```python
# 데이터 수집

data=pd.read_csv("./yemoonsaBigdata/datasets/Part2/housing_data.csv",header=None,sep=',')
col_names=['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT','MEDV','isHighValue']
data.columns=col_names
data
```

```python
# 결측치가 'CRIM'변수에 20개나 있는 것을 확인
data.info()
```

```python
# 결측치가 있는 행 20개를 제거
data=data.loc[data['CRIM'].notnull(),]
data
```

```python
X_cols=['LSTAT','PTRATIO','TAX','AGE','NOX','INDUS','CRIM'] # 독립변수 지정

# 독립변수들과 이에 대응되는 값
X = data[X_cols].values  

# 종속변수 (예측하고자 하는 변수)
Y = data['MEDV'].values
```

```python
# 학습용 데이터와 테스트용 데이터로 분할
X_train,X_test,Y_train,Y_test = train_test_split(X, Y, test_size=0.3, random_state = 123)
```

```python
# 독립변수 중 수치형 데이터 스케일링
scaler = MinMaxScaler()

X_train_r_scaled = scaler.fit_transform(X_train)
X_test_r_scaled = scaler.transform(X_test)
```

```python
# 모델 정의 및 구축(학습시키는 것을 의미)

model_lr = LinearRegression()
model_lr.fit(X_train_r_scaled, Y_train)
```

```python
# 각 입력변수에 대한 회귀계수를 출력
print(model_lr.coef_)
```

```python
# y절편 출력
print(model_lr.intercept_)
```

```python
# 예측값 생성
Y_pred = model_lr.predict(X_test_r_scaled)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_absolute_percentage_error
# 모형 평가 >> MAE, MSE, MAPE

print("선형 회귀 결과")
print("MAE : %.3f" % mean_absolute_error(Y_test, Y_pred))
print("MSE : %.3f" % mean_squared_error(Y_test,Y_pred))
print("MAPE : %.3f" % mean_absolute_percentage_error(Y_test,Y_pred))
```
